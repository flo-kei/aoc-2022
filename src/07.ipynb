````python
from dataclasses import dataclass
import re


@dataclass
class File:
    name: str
    size: int


class Directory:
    name: str
    files: list[File]
    childDirectories: list
    parentFolder: any
    size: int

    def __init__(self) -> None:
        self.name = ""
        self.childDirectories = []
        self.files = []
        self.parentFolder = None
        self.size = 0

    def __repr__(self) -> str:
        return (f"Folder(name={self.name},files={self.files},folders={self.childDirectories}, parentFolder={self.parentFolder})")


rootDir = Directory()
rootDir.name = '/'
currentDir = rootDir

path = ''


def print_directories(dir: Directory, depth: int):
    print(' '*(depth) + '- '+dir.name)
    for childDir in dir.childDirectories:
        print_directories(childDir, depth+1)
    for file in dir.files:
        print(' '*(depth+2) + '- '+file.name, file.size)


def calculate_size_of_dir(dir: Directory) -> int:
    for childDirectory in dir.childDirectories:
        dir.size += calculate_size_of_dir(childDirectory)

    for file in dir.files:
        dir.size += file.size

    return dir.size


def get_child_dir(dirName: str, dir: Directory):
    for child in currentDir.childDirectories:
        if child.name == dirName:
            return child

    return None


def find_directories_with_max_size(maxSize: int, dir: Directory):
    matching_ones = []

    for child in dir.childDirectories:
        recur_matches = find_directories_with_max_size(maxSize, child)
        for match in recur_matches:
            matching_ones.append(match)
    if dir.size <= maxSize:
        matching_ones.append(dir)

    return matching_ones


with open('input', 'r') as f:
    for line in f:
        data = re.findall('\\S+', line)
        # print(data)
        if data[0] == '$':
            # command
            if data[1] == 'cd':
                if data[2] == '..':
                    path = path[:path.rindex('/')]
                    if currentDir.parentFolder != None:
                        currentDir = currentDir.parentFolder
                elif data[2] == '/':
                    path = data[2]
                    currentDir = rootDir
                else:
                    path = path + '/' + data[2]
                    childDir = get_child_dir(data[2], currentDir)
                    if childDir == None:
                        childDir = Directory()
                        childDir.name = data[2]
                        childDir.parentFolder = currentDir
                        currentDir.childDirectories.append(childDir)
                    currentDir = childDir

        else:
            if data[0] == 'dir':
                childDir = get_child_dir(data[1], currentDir)
                if childDir == None:
                    subDir = Directory()
                    subDir.name = data[1]
                    subDir.parentFolder = currentDir
                    currentDir.childDirectories.append(subDir)
            else:
                file = File(data[1], int(data[0]))
                currentDir.files.append(file)

#print_directories(rootDir, 0)
currentDir = rootDir
calculate_size_of_dir(rootDir)

print('rootdir size:',rootDir.size)


matches = find_directories_with_max_size(100000, rootDir)

totalSizeOfMatchingDirectories = 0
for x in matches:
    totalSizeOfMatchingDirectories += x.size

print('totalsize:',totalSizeOfMatchingDirectories)
````

````python
from dataclasses import dataclass
import re


@dataclass
class File:
    name: str
    size: int


class Directory:
    name: str
    files: list[File]
    childDirectories: list
    parentFolder: any
    size: int

    def __init__(self) -> None:
        self.name = ""
        self.childDirectories = []
        self.files = []
        self.parentFolder = None
        self.size = 0

    def __repr__(self) -> str:
        return (f"Folder(name={self.name},files={self.files},folders={self.childDirectories}, parentFolder={self.parentFolder})")


rootDir = Directory()
rootDir.name = '/'
currentDir = rootDir

path = ''


def print_directories(dir: Directory, depth: int):
    print(' '*(depth) + '- '+dir.name)
    for childDir in dir.childDirectories:
        print_directories(childDir, depth+1)
    for file in dir.files:
        print(' '*(depth+2) + '- '+file.name, file.size)


def calculate_size_of_dir(dir: Directory) -> int:
    for childDirectory in dir.childDirectories:
        dir.size += calculate_size_of_dir(childDirectory)

    for file in dir.files:
        dir.size += file.size

    return dir.size


def get_child_dir(dirName: str, dir: Directory):
    for child in currentDir.childDirectories:
        if child.name == dirName:
            return child

    return None


def find_directories_with_max_size(maxSize: int, dir: Directory):
    matching_ones = []

    for child in dir.childDirectories:
        recur_matches = find_directories_with_max_size(maxSize, child)
        for match in recur_matches:
            matching_ones.append(match)
    if dir.size <= maxSize:
        matching_ones.append(dir)

    return matching_ones


def find_directories_with_min_size(minSize: int, dir: Directory):
    matching_ones = []

    for child in dir.childDirectories:
        recur_matches = find_directories_with_min_size(minSize, child)
        for match in recur_matches:
            matching_ones.append(match)
    if dir.size >= minSize:
        matching_ones.append(dir)

    return matching_ones



with open('input', 'r') as f:
    for line in f:
        data = re.findall('\\S+', line)
        # print(data)
        if data[0] == '$':
            # command
            if data[1] == 'cd':
                if data[2] == '..':
                    path = path[:path.rindex('/')]
                    if currentDir.parentFolder != None:
                        currentDir = currentDir.parentFolder
                elif data[2] == '/':
                    path = data[2]
                    currentDir = rootDir
                else:
                    path = path + '/' + data[2]
                    childDir = get_child_dir(data[2], currentDir)
                    if childDir == None:
                        childDir = Directory()
                        childDir.name = data[2]
                        childDir.parentFolder = currentDir
                        currentDir.childDirectories.append(childDir)
                    currentDir = childDir

        else:
            if data[0] == 'dir':
                childDir = get_child_dir(data[1], currentDir)
                if childDir == None:
                    subDir = Directory()
                    subDir.name = data[1]
                    subDir.parentFolder = currentDir
                    currentDir.childDirectories.append(subDir)
            else:
                file = File(data[1], int(data[0]))
                currentDir.files.append(file)

#print_directories(rootDir, 0)
currentDir = rootDir
print('\n\n')
calculate_size_of_dir(rootDir)

print('rootdir size:',rootDir.size)


#matches = find_directories_with_min_size(8381165, rootDir)
matches = find_directories_with_min_size(1072511, rootDir)

matches = sorted(matches, key=lambda m: m.size)

for x in matches[:3]:
    print(x.name,x.size)

print('\nselected-folder:')
print(matches[0].name)

print('\nselected folder size:')
print(matches[0].size)
````
